```python
import tensorflow as tf
import numpy as np
```

```python
def dynamic_rnn(rnn_type='lstm'):
    # 创建输入数据,7代表batch size,6代表输入序列的最大步长(max time),4代表每个序列的维度
    X = np.random.randn(7, 6, 4)
 
    # 第二个输入的实际长度为4
    X[1, 4:] = 0
 
    #记录三个输入的实际步长
    X_lengths = [6, 4, 6]
 
    rnn_hidden_size = 5
    if rnn_type == 'lstm':
        cell = tf.compat.v1.nn.rnn_cell.BasicLSTMCell(num_units=rnn_hidden_size, state_is_tuple=True)
    else:
        cell = tf.compat.v1.nn.rnn_cell.GRUCell(num_units=rnn_hidden_size)
 
    outputs, last_states = tf.compat.v1.nn.dynamic_rnn(
        cell=cell,
        dtype=tf.float64,
        #sequence_length=X_lengths,
        inputs=X)
 
    print( ' outputs：')
    print( outputs.numpy() )
    print( 'last_states: ')
    print( last_states.numpy())
    with tf.compat.v1.Session() as session:
        session.run(tf.compat.v1.global_variables_initializer())
        o1, s1 = session.run([outputs, last_states])
        print(np.shape(o1))
        print(o1)
        print(np.shape(s1))
        print(s1)
 
if __name__ == '__main__':
    #tf.compat.v1.disable_eager_execution()
    dynamic_rnn(rnn_type='lstm')
```
